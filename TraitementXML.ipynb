```python
import lxml.etree as et
```

```python
doc_sport = et.parse('sport.xml')
doc_sport
```

```python
schema_sport = et.parse('sports.xsd')
schema_sport
```

```python
xsd_sport = et.XMLSchema(schema_sport)
xsd_sport
```

```python
parser = et.XMLParser(schema=xsd_sport)
parser
```

```python
doc_sport = et.parse('sport.xml', parser=parser)
doc_sport
```
